```python
df = pd.read_excel('../data/words.xlsx', header=None)
df = df[[0, 1]]
df
```

```python
df_tmp = pd.read_excel('../data/tieng_hmong_1.xlsx', header=None)
df_tmp = df_tmp[[0, 1]]
df = pd.concat([df, df_tmp])
df
```

```python
df_tmp = pd.read_excel('../data/tieng_hmong_2.xlsx', header=None)
df_tmp = df_tmp[[0, 1]]
df = pd.concat([df, df_tmp])
df
```

```python
df = df.rename(columns={0: 'mong', 1: 'vietnamese'})
df
```

```python
df.info()
```

```python
def normalize(text):
    text = text.strip()
    text = text.strip('.').strip(',').strip(':').strip(';').strip('?').strip('!').strip('"').strip("'")
    text = text.replace("'", "").replace('"', '').replace('“', '').replace('”', '').replace('‘', '').replace('’', '')
    text = text.lower()
    return text

df.dropna(inplace=True)
df['mong'] = df['mong'].apply(normalize)
df['vietnamese'] = df['vietnamese'].apply(normalize)
df
```

```python
df.to_csv('../data/dataset.csv', index=False)
```
